```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
text_data_1="""Matsya: The fish avatar. 
King Vaivasvata Manu finds a little fish in the palm of his hands when performing the tarpana (water-offering). 
The fish asks Manu if his riches and power was enough to give the fish a nice home. 
Manu keeps the fish to give it a home, but the fish keeps expanding, which breaks Manu's pride about his wealth. 
Eventually, he releases it into the ocean, realizing it is Vishnu himself. 
Vishnu informs Manu of the coming destruction of the world, 
by means of fires and floods, and directs Manu to collect "all creatures of the world" 
and keep them safe on a boat built by the gods. 
When the deluge (Pralaya) occurs, Vishnu appears as a great fish with a horn,
to which Manu ties the boat, which leads them into safety"""
```

```python
# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, max_words=10000, background_color='white').generate(text_data_1)

# Generate the word cloud image
wordcloud_image = wordcloud.to_image()
```

```python
# Display the word cloud using matplotlib
plt.imshow(wordcloud_image, interpolation='bilinear')
plt.axis('off')
plt.title("MATSYA: The fish avtar",fontsize = 25, color = "magenta")
plt.show()
```
